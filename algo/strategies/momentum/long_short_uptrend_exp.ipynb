```python
import numpy as np
import pandas as pd
import yfinance as yf
from datetime import date, datetime, timedelta

from indicators.moving_averages import simple_moving_average
```

```python
# TODO: explore with hourly.


# Instruments
# ticker = "SPY"
ticker = "CVX"  # Chevron
interval = "1d"

# Dates
end_date = date.today()
start_date = end_date - timedelta(days=365)

df = yf.download(ticker, start=start_date.strftime("%Y-%m-%d"), end=end_date.strftime("%Y-%m-%d"), interval=interval)

price_col_raw = "Adj Close"
price = "price"

df.rename(columns={price_col_raw: price}, inplace=True)
df.head()
```

```python
# Double [[]] keeps df as DataFrame, not Series.
# df = df[["Date", price]].dropna().reset_index(drop=True)
df = df[[price]].dropna()  #.reset_index()
df.head()
```

```python
w = 200
x = df[price]

ma = np.convolve(x, np.ones(w), "valid") / w
ma
```

```python
span = 200
sma_200 = f"sma_{span}_{price}"
df = df.pipe(simple_moving_average, colname=price, span=span)

span = 10
sma_10 = f"sma_{span}_{price}"
df = df.pipe(simple_moving_average, colname=price, span=span)
```
